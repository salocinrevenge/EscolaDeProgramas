```python
import plotly.graph_objects as go
import numpy as np
import umap.umap_ as umap
import pickle
```

```python
def obter_dados(nome_arquivo):
    with open(nome_arquivo) as f:
        data = f.read().splitlines()
        data = [i.split(",") for i in data[1:]]
        rotulo = [int(i[-1]) for i in data]
        data = [i[:360] for i in data]
        data = [[float(j) for j in i] for i in data]

    return np.array(data), np.array(rotulo)

def mostrar_dados(data, rotulo, modelo):
    
    try:
        with open(modelo,"rb") as f:
            reducer = pickle.load(f)
    except:
        print("Treinando um novo modelo UMAP")
        reducer = umap.UMAP()
        reducer.fit(data)
        with open(modelo,"wb") as f:
            pickle.dump(reducer,f)
    embedding = reducer.transform(data)


    #plota os valores de treino em um scatter, com cores diferentes para cada classe, de forma que seja possivel desligar as cores
    fig = go.Figure()
    for i in range(0,6):
        fig.add_trace(go.Scatter(x=embedding[rotulo==i][:,0], y=embedding[rotulo==i][:,1], mode='markers', name=f"Classe {i}",marker=dict(size=15, line=dict(width=4, color='DarkSlateGrey'))))
    fig.show()


mostrar_dados(*obter_dados("data/kuhar/train.csv"),"modelos/umap.pk2")

```
